```python
from paddleocr import PaddleOCR
import matplotlib.pyplot as plt
from PIL import Image

IMG_PATH_ppocrv3 = "experiments/runs/img_runs/MD1.jpg"
MODEL_DIR_ppocrv3 = "experiments/models/svtr/trainSVTRLCNet_paddle_latinppocrv3/export_model"
DICT_DIR_ppocrv3 = "experiments/models/svtr/trainSVTRLCNet_paddle_latinppocrv3/plate_dict.txt"

ocr_ppocrv3 = PaddleOCR(
    det=False,
    cls=False,
    rec_model_dir=MODEL_DIR_ppocrv3,
    rec_char_dict_path=DICT_DIR_ppocrv3,
    #ocr_version="PP-OCRv3",
    #lang="latin",
    rec_algorithm="SVTR_LCNet",
    use_space_char=False,
    enable_mkldnn=False,
    rec_image_shape="3, 48, 320",
    use_angle_cls=False,
    use_gpu=False,
    show_log=False
)

result = ocr_ppocrv3.ocr(IMG_PATH_ppocrv3, det=False, cls=False)

print("RAW:", result)
if result and result[0]:
    print("TEXT:", result[0][0][0])

plt.imshow(Image.open(IMG_PATH_ppocrv3))
plt.show()
```
