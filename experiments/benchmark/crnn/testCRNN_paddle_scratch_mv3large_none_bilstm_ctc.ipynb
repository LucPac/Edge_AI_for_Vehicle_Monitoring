```python
from paddleocr import PaddleOCR
import os

# PATH
IMG_PATH_mv3 = "dataset/dataset/test/images"
LABELS_PATH_mv3 = "dataset/dataset/test/labels.txt"
MODEL_DIR_mv3 = "experiments/models/crnn/trainCRNN_paddle_scratch_mv3large_none_bilstm_ctc/export_model"
DICT_DIR_mv3 = "experiments/models/crnn/trainCRNN_paddle_scratch_mv3large_none_bilstm_ctc/plate_dict.txt"

# Load OCR model
ocr_mv3 = PaddleOCR(
    det=False,
    cls=False,
    rec_model_dir=MODEL_DIR_mv3,
    rec_char_dict_path=DICT_DIR_mv3,
    rec_algorithm="CRNN",
    rec_image_shape="3, 32, 320",
    use_angle_cls=False,
    use_gpu=False,
    show_log=False
)

# Split img name - label
def parser_imgname_imglabel(labels_path):
    with open(labels_path, "r", encoding="utf-8") as f:
        lines = f.readlines()
    imgname_label_dict = {}
    for line in lines:
        if not line.strip():
            continue
        imgname, label = line.strip().split("\t")
        imgname_label_dict[imgname] = label
    return imgname_label_dict

# Evaluate OCR model
def evaluate(ocr, img_dir, labels_path):
    imgname_label_dict = parser_imgname_imglabel(labels_path)
    total_samples = len(imgname_label_dict)
    correct_predictions = 0

    for imgname, true_label in imgname_label_dict.items():
        img_path = os.path.join(img_dir, imgname)

        result = ocr.ocr(img_path, det=False, cls=False)

        predicted_label = ""
        if result and result[0]:
            try:
                predicted_label = result[0][0][0]
            except IndexError:
                predicted_label = ""

        print(f"{imgname:<25} {true_label:<15} - {predicted_label}")

        if predicted_label == true_label:
            correct_predictions += 1

    accuracy = correct_predictions / total_samples
    return accuracy

accuracy_mv3 = evaluate(ocr_mv3, IMG_PATH_mv3, LABELS_PATH_mv3)
print(f"Accuracy: {accuracy_mv3}")
```
